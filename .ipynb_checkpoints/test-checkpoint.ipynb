```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def introduction(p, S):
    r = np.random.rand(S)
    r[r < p] = 0
    return r
    
```

```python
np.random.seed(1234)

# Set parameters
S = 100
x0 = np.random.rand(S) * 10
n = 1000

# Simulate exponential growth in continuous time limit
x = np.zeros((n,S))
x[0,:] = x0
for i in range(1,n):
    gamma = np.abs(1 + np.random.normal(size=(S)))
    x[i,:] = gamma * x[i-1,:] + introduction(0.9, S)

log_x = np.log(x)
```

```python
plt.plot(log_x[:,0])
plt.plot(log_x[:,1])
plt.plot(log_x[:,2])
```

```julia
using DataFrames, DataFramesMeta, GLM
using Statistics, StatsBase
using FFTW
using NFFT
using CSV
using Plots
using Random, Distributions
```

```julia
# Import synthetic data
synth_data = CSV.read("./Data/synthetic/SLM.csv", DataFrame)
first(synth_data, 5)
```

```julia
# Consider only steady state regime
# I also consider only a portion of it since using all of the samples makes the computation heavier without
# having any difference in the final result
df = synth_data[(synth_data.time .> 5000) .& (synth_data.time .< 7000),:];

# Get the variables that are always non-zero
nonzero_vars = combine(groupby(df, :variable)) do subdf
    (; variable = subdf.variable[1], all_nonzero = all(>(0), subdf.value))
end

# Filter the variables where all values are > 0
allowed = nonzero_vars.variable[nonzero_vars.all_nonzero .== true]

# Filter the original DataFrame
df_filtered = filter(row -> row.variable in allowed, df);
```

# PSD analysis using FFT

```julia
# Set paramas and variables to perform FFT
N = length(unique(df_filtered.time))
otu_ids = unique(df_filtered.variable) # Get otu_ids
t_indices = unique(df_filtered.time .- minimum(df_filtered.time))
Δt = t_indices[2] - t_indices[1]
fs = 1 / Δt  # sampling rate is 1 day
Nf = fs / 2.0 # Nyquist frequency
frequencies =  fftshift(fftfreq(N, fs)) # Frequency domain

otu_count = 0 # Needed for mean PSD
mean_S = zeros(N)  # initialize array
for otu in otu_ids
    x = log.(collect(filter(row -> row.variable == otu, df_filtered).value))
    x .-= mean(x)

    # Some species remains at very low abundance so their fluctuations are essentially null. This means that their signal 
    # is seen as constant which means that all the frequencies are almost null and so is their power spectrum.
    if std(x) < 1e-10
        println("Skipping $otu — near-constant signal")
        continue
    end

    otu_count += 1

    # Compute FFT
    fhat = fftshift(fft(x))
    S = abs2.(fhat) .* (Δt / (2*N + 1))
    mean_S .+= S
end

# Take only positive frequencies
positive = frequencies .> 0
frequencies = frequencies[positive]
mean_S = mean_S[positive] ./ otu_count;
```

```julia
# Filter frequencies to avoid bias at too low and too high frequencies
# Essentially the fact the we have linear frequency binning but we compute the slope in log-log space creates problems
# at very low freqs because there are less points and at very high freqs because there are too many points. Moreover these
# methods are exact only in the infinite time case so we need to consider border effects.
mask = (frequencies .> 10 * (1 / N)) .& (frequencies .< maximum(frequencies) / 10)
frequencies = frequencies[mask]
mean_S = mean_S[mask]

# Take logs
log_f = log10.(frequencies)
log_S = log10.(mean_S)

# Put into a DataFrame and fit linear model: log_S ~ log_f
plot_df = DataFrame(log_f=log_f, log_S=log_S)
model = lm(@formula(log_S ~ log_f), plot_df)

# Extract the slope and intercept
coeffs = coef(model)
slope = coeffs[2]
intercept = coeffs[1]

println("Estimated slope: ", slope)
println("Intercept: ", intercept)

# Plot results
plot(log_f, log_S, label=:none, xlabel="log₁₀(frequency)", ylabel="log₁₀(power)", legend=:bottomleft,
    title="Mean PWD of SLM")
plot!(log_f, predict(model), label="Fit: slope = $(round(slope, digits=2))", lw=2)
```

# PSD analysis using NFFT

```julia
# Set paramas and variables to perform NFFT
N = length(unique(df_filtered.time))
otu_ids = unique(df_filtered.variable) # Get otu_ids
t_indices = unique(df_filtered.time .- minimum(df_filtered.time))
Δt = t_indices[2] - t_indices[1]
fs = 1 / Δt  # sampling rate is 1 day
Nf = fs / 2.0 # Since sample rate is 1 day
frequencies = (-Int(N/2):Int(N/2 - 1)) * fs / N # Frequency domain

println("Window length:  $N")

mean_S = zeros(N) # Initialize array
otu_count = 0 # Needed for normalization
otu_ids = unique(df_filtered.variable) # Get otu_ids
for otu in otu_ids
    t_normalized = t_indices ./ N .- 0.5  # The algorithm work for t ∈ [-0.5, 0.5)
    x = log.(collect(filter(row -> row.variable == otu, df_filtered).value))
    x .-= mean(x) # Detrend signal to avoid peak at zero frequency and have comparable signals
    
    # Compute non-uniform FFT only for a 'sufficient' number of samples
    otu_count += 1
    p_nfft = NFFT.plan_nfft(t_normalized, N, reltol=1e-9)
    fhat = adjoint(p_nfft) * x

    # Compute normalized power spectrum density (periodogram)
    S = abs2.(fhat) .* (Δt / (2*N + 1))
    mean_S .+= S # Mean PSD: this step should give a more accurate result for the PSD supposing that all trajectories are equivalent
end

println("OTUs considered:  $otu_count")

# Take only positive frequencies
positive = frequencies .> 0
frequencies = frequencies[positive]
mean_S = mean_S[positive] ./ otu_count;
```

```julia
# Filter frequencies to avoid bias at too low and too high frequencies
# Essentially the fact the we have linear frequency binning but we compute the slope in log-log space creates problems
# at very low freqs because there are less points and at very high freqs because there are too many points. Moreover these
# methods are exact only in the infinite time case so we need to consider border effects.
mask = (frequencies .> 100 * (1 / N)) .& (frequencies .< maximum(frequencies) / 10)
frequencies = frequencies[mask]
mean_S = mean_S[mask]

# Take logs
log_f = log10.(frequencies)
log_S = log10.(mean_S)

# Put into a DataFrame and fit linear model: log_S ~ log_f
plot_df = DataFrame(log_f=log_f, log_S=log_S)
model = lm(@formula(log_S ~ log_f), plot_df)

# Extract the slope and intercept
coeffs = coef(model)
slope = coeffs[2]
intercept = coeffs[1]

println("Estimated slope: ", slope)
println("Intercept: ", intercept)

# Plot results
plot(log_f, log_S, label=:none, xlabel="log₁₀(frequency)", ylabel="log₁₀(power)", legend=:bottomleft,
    title="Mean PWD of SLM")
plot!(log_f, predict(model), label="Fit: slope = $(round(slope, digits=2))", lw=2)
```
